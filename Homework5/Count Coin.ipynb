```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('sIMG_8253.JPG')
edges1 = cv2.Canny(img,100,120)
blur = cv2.GaussianBlur(img,(5,5),0)
edges2 = cv2.Canny(blur,100,120)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges2,cmap = 'gray')
plt.title('Edge Image with smooth'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
im = cv2.imread('sIMG_8256.JPG')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(im,(5,5),0)
edges2 = cv2.Canny(blur,100,120)
ret,thresh = cv2.threshold(edges2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
im = cv2.drawContours(im, contours, -1, (255,0,0), 3)
```

```python
plt.figure(figsize=(12,9))
plt.subplot(121),plt.imshow(im,cmap = 'gray')
plt.title('im'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.show()
```
