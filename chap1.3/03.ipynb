```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = np.zeros((400,512,3), np.uint8)
for i in range(255):
    img[:,i]=i
    img[:,256+i]=255-i
```

```python
plt.imshow(img)
plt.show()
```

```python
ret,mask = cv2.threshold(img,230,80,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(mask)
plt.show()
```

```python
mask_inv = cv2.bitwise_not(mask)
```

```python
plt.imshow(mask_inv)
plt.show()
```
