## Method - 2 : Using Matplotlib

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('sea.jpg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
hist = cv2.calcHist( [hsv], [0, 1], None, [180, 256], [0, 180, 0, 256] )
```

```python
plt.imshow(hist,interpolation = 'nearest')
plt.show()
```
