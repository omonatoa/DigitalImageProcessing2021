```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('y.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
kernel = np.ones((9,9),np.uint8)
erosion2 = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.figure(figsize=(32,20))
plt.subplot(191),plt.imshow(img,'gray'),plt.title('Input')
plt.subplot(192),plt.imshow(erosion,'gray'),plt.title('erosion')
plt.subplot(193),plt.imshow(erosion2,'gray'),plt.title('erosion2')
plt.subplot(194),plt.imshow(dilation,'gray'),plt.title('dilation')
plt.subplot(195),plt.imshow(opening,'gray'),plt.title('opening')
plt.subplot(196),plt.imshow(closing,'gray'),plt.title('closing')
plt.subplot(197),plt.imshow(gradient,'gray'),plt.title('gradient')
plt.subplot(198),plt.imshow(tophat,'gray'),plt.title('tophat')
plt.subplot(199),plt.imshow(blackhat,'gray'),plt.title('blackhat')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('fp2.jpg',0)
blur = cv2.GaussianBlur(img,(3,3),0)
ret3,th3 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
opening = cv2.morphologyEx(th3,cv2.MORPH_OPEN,kernel)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(131),plt.imshow(img,'gray'),plt.title('Input')
plt.subplot(132),plt.imshow(th3,'gray'),plt.title('Output')
plt.subplot(133),plt.imshow(opening,'gray'),plt.title('Output')
plt.show()
```
